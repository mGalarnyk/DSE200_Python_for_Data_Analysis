```python
# (3) same as 2, but instead of single words, count the number of occurances of **word pairs** i.e.
# ('bands','which'),('which','have'),...
```

```python
for char in '-.,\n':
    Text=Text.replace(char,' ')
Text = Text.lower()

tokens = Text.split(" ")
token_count = {}
last_token  = None

for token in tokens:
    if token == "":
        continue
    if last_token == None:
        last_token = token
        continue
    bigram = (last_token,token)
    if bigram not in token_count:
        token_count[bigram] = 0
    token_count[bigram] += 1
    last_token = token

sorted(token_count.items(), key = lambda x: x[1], reverse = True)
```

(4) For the list `l =  [ 6, -5, 3, -2, 0, 8, -9, -7 ]`, return the numbers which are greater than the average (of these numbers)

```python
l = [ 6, -5, 3, -2, 0, 8, -9, -7 ]
mean_l = sum([x/len(l) for x in l])
above_average_l = [x for x in l if x > mean_l]
(mean_l, above_average_l)
```

(5) for the tuples (2, 5, 3, 7, 10, 20, 54, 89) and (15, 16, 3, 37, 54, 21, 20, 5) return a sorted tuple that contains
only the elements common to both the tuples (use sets)

```python
xs = (2, 5, 3, 7, 10, 20, 54, 89)
ys = (15, 16, 3, 37, 54, 21, 20, 5)

xs_set = set(xs)
ys_set = set(ys)

union_set = xs_set.intersection(ys_set)
tuple(sorted(union_set))
```

(12) write a program to find solutions of the 8 queen problem.
>The eight queens problem is the problem of placing eight queens on an 8Ã—8 chessboard such that none of them attack one >another (no two are in the same row, column, or diagonal). Below is one of the arrangements of an 8 queen solution

<img src="8queen.jpg">

```python
def under_attack(col, queens):
    left = right = col

    for r, c in reversed(queens):
        left, right = left - 1, right + 1

        if c in (left, col, right):
            return True
    return False

def solve(n):
    if n == 0:
        return [[]]

    smaller_solutions = solve(n - 1)

    return [solution+[(n,i+1)]
        for i in xrange(8)
            for solution in smaller_solutions
                if not under_attack(i+1, solution)]

for answer in solve(8):
    print [r for (c,r) in answer]
```

(13) Write a program to make a list of all prime numbers less than 3000. A prime number can only be divided evenly by 1 and itself.

```python
primes = []
candidates = range(2,3000)
while len(candidates) > 0:
    prime = candidates[0]
    primes.append(prime)
    candidates = candidates[1:]
    for x in candidates:
        if (x%prime) == 0:
            candidates.remove(x)
print primes
```

(14) Solve the above problem using Set instead of List 

```python
primes = []
candidates = set(range(2,3000))
while len(candidates) > 0:
    prime = min(candidates)
    primes.append(prime)
    candidates.remove(prime)
    dropSet = set()
    for x in candidates:
        if (x%prime) == 0:
            dropSet.add(x)
    candidates = candidates.difference(dropSet)
print primes
```
